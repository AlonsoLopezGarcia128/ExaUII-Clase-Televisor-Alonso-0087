#### Este código implementa la Programación Orientada a Objetos para gestionar información de televisores mediante una clase que define tres atributos específicos: marca, pulgadas y tecnología. El programa funciona creando un molde que permite al usuario ingresar datos de forma interactiva a través del teclado, utilizando la función int() para convertir el valor de las pulgadas en un número entero. Una vez capturada la información, el código utiliza una función de salida para imprimir los detalles organizadamente en la pantalla. Al final, el proceso se repite manualmente para dos objetos distintos, llamados tv1 y tv2, logrando registrar y visualizar la información de ambos dispositivos de manera lineal y sencilla sin necesidad de utilizar estructuras de control complejas como ciclos o condiciones.

![texto del vínculo](https://raw.githubusercontent.com/AlonsoLopezGarcia128/ExaUII-Clase-Televisor-Alonso-0087/refs/heads/main/infoexaIIclasstelevisor.png)

```python
#### Alonso Lopez Garcia NL = 31   Fecha = 30/Abril/2026

class Television:
    def __init__(self, marca, pulgadas, tecnologia):
        self.marca = marca
        self.pulgadas = pulgadas
        self.tecnologia = tecnologia

    def leerdatos(self):
        # Usamos input estándar y convertimos pulgadas a entero (int)
        self.marca = input("Ingrese la marca de la TV: ")
        self.pulgadas = int(input("Ingrese las pulgadas (solo números): "))
        self.tecnologia = input("Ingrese la tecnología (LED, OLED, 4K, etc.): ")

    def mostrardatos(self):
        print("--- Datos de la Televisión ---")
        print("Marca:     ", self.marca)
        print("Pulgadas:  ", self.pulgadas, '"')
        print("Tecnología:", self.tecnologia)

# --- Creación de los objetos tv1 y tv2 ---

print("Registro de la TV Numero 1")
tv1 = Television("", 0, "") # Inicialización vacía
tv1.leerdatos()
tv1.mostrardatos()
print("*********************************")

print("Registro de la TV Numero 2")
tv2 = Television("", 0, "") # Inicialización vacía
tv2.leerdatos()
tv2.mostrardatos()
print("*********************************")

print("Programa realizado por Alonso Lopez Garcia NL = 31")
```
